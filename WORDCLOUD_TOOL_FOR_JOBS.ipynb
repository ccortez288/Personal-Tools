```python
import requests, json, wordcloud
import matplotlib.pyplot as plt
def cloudCreator(text):
  cloud = wordcloud.WordCloud()
  cloud.generate(text)
  plt.imshow(cloud, interpolation = 'bilinear')
  plt.axis('off')
  plt.show()
  plt.close()
```

```python
#Copy paste job description here and a wordcloud will be creatred for it so that you can see the keywords in it. 
#Use to tailor your resume as such to match the keywords in the job description

input_text = input('Input job description')

cloudCreator(input_text)
```

```python
#Input your resume to see the words that pop up in your resume, use it to see how you should adjust
#to the job you are applying for
input_resume = input('Input resume')

cloudCreator(input_resume)
```
